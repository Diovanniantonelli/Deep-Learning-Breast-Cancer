```python
import pandas as pd
```

```python
X = pd.read_csv('/content/entradas_breast.csv')
X
```

```python
y = pd.read_csv('/content/saidas_breast.csv')
y
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X_training, X_test, y_training, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_training.shape, y_training.shape
```

```python
X_test.shape, y_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
rede_neural = Sequential([
    tf.keras.layers.InputLayer(input_shape=(30,)), #camada de entrada
    tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer= 'random_uniform'),
      tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer= 'random_uniform'),#camada oculta, Dense = conecta todos os neuronios
    tf.keras.layers.Dense(units=1, activation='sigmoid') #units = quantidade de neuronios (30 + 1)/2 -> entrada e saida
])
```

```python
rede_neural.summary()
```

```python
otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, decay=0.0001, clipvalue=0.5)
```

```python
rede_neural.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy'])
```

```python
rede_neural.fit(X_training, y_training, batch_size=10, epochs=100) #batch -> grupos de 10 em 10
```

```python
pesos0 = rede_neural.layers[0].get_weights()
pesos0
```

```python
len(pesos0)
```

```python
len(pesos0[0])
```

```python
previsoes = rede_neural.predict(X_test)
previsoes = previsoes > 0.5
```

```python
previsoes
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy_score(y_test, previsoes)
```

```python
confusion_matrix(y_test, previsoes)
```

```python
rede_neural.evaluate(X_test, y_test) #DEIXAR SO UMA CAMADA PARA PROBLEMAS MAIS SIMPLES
```
